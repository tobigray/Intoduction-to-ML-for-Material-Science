Advanced exercise: calculate Young's modulus

Young's modulus is given by:

E=9KG/(G+3K) 

where 
 K is the bulk modulus (column K_VRH), and 
 G is the shear modulus (column G_VRH).

Calculate Young's modulus for all entries in the dataset and store them in a new column called E_VRH. What is the average Young modulus over the entire dataset?

```python
et["E_VRH"] = (9 * et["K_VRH"] * et["G_VRH"])/((et["G_VRH"])+(3*et["K_VRH"]))

et[["E_VRH"]]

mean_EVRH = et["E_VRH"].mean()

print(mean_EVRH)
```

Featurizers - transforming material parameters into machine learnable features, 
or they convert pymatgen objects to into vectors and matrices that can be used in dataframes. 
You can convert composition based on chemistry, crystal structures based on geometry and chemistry, density of states, and band structures. Check matminer website for the 

```python
from pymatgen.core import Composition #First,  we import composition form pymatgen, since this is what we want to turn into machine learnable vector 


fe2o3 = Composition("Fe2O3") #Then, we make a new variable for a fe2O3 compound, this is a trivial example


from matminer.featurizers.composition.element import ElementFraction #Then, we get the Element fraction using the inbuilt matminer element fraction featurizer


eo = ElementFraction() #This will simply tell us the fractions of elements in a composition, it doesnt need an argument which is why the () is empty. 
#The variable 'eo' thus represents the imported element fraction
```

```python
element_fractions = eo.featurize(fe2o3) # This gives the list vector equivalents of the compound from pymatgen

print(element_fractions)
```

```python
element_fraction_labels = eo.feature_labels() # This shows the respective elements represented in the periodic table as shown by the vectors above

print(element_fraction_labels)
```

```python
print(element_fraction_labels[7], element_fractions[7]) #Now, lets view only the elements represented here, Oxygen is the 8th element, so we can access it with 7 in python as counting starts from 0

print(element_fraction_labels[25], element_fractions[25]) # same idea goes for Fe, the 26th element 
```

```python
#This is to find the file in our resources directory

import os
from matminer.utils.io import load_dataframe_from_json

# Specifies the full path to your JSON file on your local pc
#matminer featurizers when put in a dataframe can be serialized into a 'json' format that can be transferred between pc's, uploaded online, etc..
json_file_path = os.path.join("C:\\Users\\Lenovo\\Downloads\\New", "elastic_tensor_2015_featurized.json")


#Remember you can also load a dataset from matminer's, or Materials Project, or Citrine's database IN A Pandas database format using the code below; then featurize it yourself
#from matminer.data_retrieval.retrieve_Citrine import CitrineDataRetrieval
#from matminer.data_retrieval.retrieve_MP import MPDataRetrieval
#df_citrine = CitrineDataRetrieval().get_dataframe(criteria='Si', properties=['band_gap'])
#df_mp = MPDataRetrieval().get_dataframe(criteria='Si', properties=['band_gap'])


# Load the DataFrame from your JSON file in my local downloads folder
etf = load_dataframe_from_json(json_file_path)

# Display the first 5 rows of the DataFrame
etf.head()
```

```python
#To make a model for predicting bulk modulus, 
# we firstly need to have our target values in a format scikit learn can understand
#You can parse in panda seires, i.e, just the columns if you want,
# But in this case, we will be getting the numpy array out of it using the ".values" attribute object

y = etf['K_VRH'].values # make a called variable called "y", set it to our etf dataframe, index it according to our target values "K_VRH"

print(y) # This prints our target values
```

```python
#Now, we also need to get our training features as we are selecting a matrix and not a single vector


X = etf.drop(["structure", "formula", "composition", "K_VRH"], axis=1) 

# Make a new variable x which is our training features, 
# and drop unnecessary columns which include the target value K_VRH as we are trying to map the learning features we generated onto our target values "K_VRH" 
# We specify axis=1 to make sure we are dropping the columns
```

```python
print("There are {} possible descriptors:".format(len(X.columns))) # This gives the length of x columns left in the feature matrix

print(X.columns) # This will show the number of possible descriptors or features(represented by a column each)
```

```python
# Now we will train our model using scikit learn and the features we have made ready

# And we will be training a RandomForestRegressor model here

# We will first train on the data to evaluate the training error, then run a cross validation where we estimate the error more accurately
```

```python
from sklearn.ensemble import RandomForestRegressor 

rf = RandomForestRegressor(n_estimators=100, random_state=1) 

# The n estimator parameter above is an example of a hyper parameter in a ML model, 
# so you set it before fitting your model
# You might have to try a bunch of parameters before getting the optimal estimation

# the second argument, random state = 1 is to make sure we all get the same answers 

RandomForestRegressor?
```

```python
# from sklearn.neural_network import MLPClassifier .. we can also use NN's for training our ML models instead of random forest regression above
# although NN is not appropriate for this, as our dataset is small
```

```python
# from sklearn.kernel_ridge import KernelRidge .. or kernel ridge regression, KRR for training our ML models instead of random forest regression above
```

```python
#from sklearn.gaussian_process import GaussianProcessRegressor .. uses GPR for training our ML model
```

```python
#from sklearn.linear_model import LinearRegression .. uses the basic LR for training our model
```

```python
rf.fit(X, y) 

# All scikit learn models have a '.fit' method which enables us to predict the target propoerty
```

```python
# Now lets see what our predictions look like, evaluate our model's performance

# Similar to fit, all scikit learn models have a predict method

y_pred = rf.predict(X)
#now we have a vector of our predicted values, this is where the actual machine learning takes place
```

```python
# now lets examine our training errors using a few things from scikit learn and numpy


import numpy as np
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y, y_pred)
print('rf training RMSE = {:.3f} GPa'.format(np.sqrt(mse))) #This is just a formatting syntax
```

```python
# Our training error is as shown above, which is not really our actual error as our goal with this model is to apply it 

#to new samples which has not yet been introduced.

#All we have done here is to train these samples and so this training RMSE way underestimates what our training error will be
```

```python
# Cross validation will much more accurately evaluate our errors

# Here, we run a number of independent experiments called folds, where we separate our training and testing data for assessing error

# So, in one fold, we might take 90% of a data for training, and keep 10% for testing

#We then repeat it K times e.g 10 times , and then we get 10 scores showing how each experiment did

# 10 is a pretty standard number for cross validation
```

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, random_state=1, shuffle=True) 

#n=10 is a standard number for cross validation, the shuffle argument will randomly shuffle the data and make sure we get same answers
```

```python
#this will give our scores on each experiments, but what we really need is the average or mean of each of these errors 

from sklearn.model_selection import cross_val_score

scores = cross_val_score(rf, X, y, scoring='neg_mean_squared_error', cv=kfold)

rmse_scores = [np.sqrt(abs(s)) for s in scores] #this is a new variable called rmse_scores with a list comprehension
print('rf Mean RMSE: {:.3f}'.format(np.mean(rmse_scores)))
```

```python
# As seen above, this almost triples the previous training error, but it's still not so bad in itself
```

```python
# Another way to evaluate our errors is by looking at some grpahs made with plotly

# First, we put some predictions on our dataframe for every fold, we use cross val function for this

from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(rf, X, y, cv=kfold) # we don't need the scoring parameter here as it only returns the raw parameters and not the scores as done above
```

```python
etf["K_VRH predicted"] = y_pred
etf["percentage_error"] = (etf["K_VRH"] - etf["K_VRH predicted"]).abs()/etf["K_VRH"] * 100

etf
```

```python
# To look at it visually, we use plotly. So I recommend you read through its documentation.
```

```python
import plotly.express as px
import plotly.graph_objects as go

reference_line = go.Scatter(
    x=[0, 400],
    y=[0, 400],
    line=dict(color="black", dash="dash"),
    mode="lines",
    showlegend=False
)

fig = px.scatter(
    etf, 
    x="K_VRH", 
    y="K_VRH predicted", 
    hover_name="formula", 
    color="percentage_error", 
    color_continuous_scale=px.colors.sequential.Bluered,
)

fig.add_trace(reference_line)
fig.show()
```

```python
# From the plot above, we can see that there are some outliers, e.g, carbon and water (colored red), we can tune the model hyper parameters and the training features to improve the model's prediction
```

```python
rf.feature_importances_ #this is the same as LASSO in ramprasad's video, to see the most important features that influenced the regression model, this is a numpy array, so we'd convert to plotly
```

```python
importances = rf.feature_importances_
included = X.columns.values
indices = np.argsort(importances)[::-1]

fig_bar = px.bar(
    x=included[indices][0:5], 
    y=importances[indices][0:5], 
    title="Feature Importances of Random Forest",
    labels={"x": "Feature", "y": "Importance"}
)
fig_bar.show()

#the results kind of intuitively makes sense as we are predicting the bulk modulus
```
